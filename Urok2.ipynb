```python
import pandas as pd

# Загружаем данные из интернета
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

print("Размер данных:", df.shape)
print("Первые 5 строк:")
print(df.head())
```

```python
import matplotlib.pyplot as plt

# Группируем данные
survival_by_class = df.groupby('Pclass')['Survived'].mean()

# Строим график
plt.bar(['1-й класс', '2-й класс', '3-й класс'],
        survival_by_class.values * 100)
plt.title('Выживаемость по классу каюты')
plt.ylabel('Процент выживших (%)')
plt.show()
```

```python
print("Пропущенные значения:")
print(df.isnull().sum())

# Пропуски в колонке Age (возраст)
age_missing = df['Age'].isnull().sum()
print(f"Возраст не указан у {age_missing} пассажиров")
```

```python
# Создаем данные с проблемами
data = {
    'student_id': [1, 2, 3, 4, 5],
    'math_score': [85, 92, 78, None, 88],  # Пропуск!
    'physics_score': [78, None, 90, 92, 88],  # Еще пропуск!
    'passed': [1, 1, 0, 1, 1]
}

df_students = pd.DataFrame(data)
```

```python
# Среднее значение по математике
mean_math = df_students['math_score'].mean()
print(f"Средняя оценка по математике: {mean_math}")

# Заполняем пропуски
df_students['math_score'].fillna(mean_math, inplace=True)

# То же самое для физики
mean_physics = df_students['physics_score'].mean()
df_students['physics_score'].fillna(mean_physics, inplace=True)

print("Данные после очистки:")
print(df_students)
```

```python
# Общий балл = математика + физика
df_students['total_score'] = df_students['math_score'] + df_students['physics_score']

# Категория успеваемости
def get_performance(total_score):
    if total_score >= 180:
        return "Отлично"
    elif total_score >= 160:
        return "Хорошо"
    else:
        return "Улучшить"

df_students['performance'] = df_students['total_score'].apply(get_performance)

print("Данные с новыми колонками:")
print(df_students)
```
